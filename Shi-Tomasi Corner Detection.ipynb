# Shi-Tomasi Corner Detection

See https://docs.opencv.org/3.2.0/d4/d8c/tutorial_py_shi_tomasi.html for more detail

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


img = cv2.imread('arrow.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

NUM_CORNER_TO_FIND = 30
corners = cv2.goodFeaturesToTrack(gray,NUM_CORNER_TO_FIND,0.01,10)

corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    # print(x, y)
    cv2.circle(img,(x,y),3,255,-1)


img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB);

plt.figure(figsize=(16, 12))
plt.imshow(img)
plt.show()
```
