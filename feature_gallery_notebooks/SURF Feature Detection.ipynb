# SURF (Speeded-Up Robust Features)

In short, SURF adds a lot of features to improve the speed in every step. Analysis shows 
* it is 3 times faster than SIFT while performance is comparable to SIFT. 
* SURF is good at handling images with blurring and rotation, 
* but not good at handling viewpoint change and illumination change.

See https://docs.opencv.org/3.2.0/df/dd2/tutorial_py_surf_intro.html for more detail

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


img = cv2.imread('pet_bottle.jpg')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hessian_th = 5000 # 400
surf = cv2.xfeatures2d.SURF_create(hessian_th)
kp = surf.detect(gray,None)
print('number of key points:', len(kp))
img=cv2.drawKeypoints(gray,kp,img, (0,0,255), 4)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB);

plt.figure(figsize=(16, 12))
plt.imshow(img)
plt.show()

kp, des = surf.detectAndCompute(img,None)
print(surf.descriptorSize())
print(des.shape)
```
