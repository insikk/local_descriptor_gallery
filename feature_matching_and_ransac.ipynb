# Demo for Feature Matching and Ransac

Requirements:
* Feature Descriptor
* Image

Return:
Side by side image with matching ransac visualization


```python
%load_ext autoreload
%autoreload 2

IMAGE_SIZE = (2, 1)

import numpy as np
import cv2
import matplotlib.pyplot as plt
from utils import matcher

import numpy as np


%matplotlib inline


img1 = cv2.imread('sample_image/kozel_query.jpg') # query image
img2 = cv2.imread('sample_image/kozel_answer.jpg')


matcher.draw_ransac(img1, img2, 
                    cv2.ORB_create(), 
                    cv2.ORB_create(), 
                    is_binary_descriptor=True, 
                    match_method='BRUTE_FORCE', 
                    description='ORB, BRTUE_FORCE, RANSAC')

matcher.draw_ransac(img1, img2, 
                    cv2.xfeatures2d.SIFT_create(), 
                    cv2.xfeatures2d.SIFT_create(), 
                    is_binary_descriptor=False, 
                    match_method='BRUTE_FORCE', 
                    description='SIFT, BRTUE_FORCE, RANSAC')

hessian_th = 400
surf = cv2.xfeatures2d.SURF_create(hessian_th)
matcher.draw_ransac(img1, img2, 
                    surf, 
                    surf, 
                    is_binary_descriptor=False, 
                    match_method='BRUTE_FORCE', 
                    description='SURF, BRTUE_FORCE, RANSAC')

star = cv2.xfeatures2d.StarDetector_create()
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
matcher.draw_ransac(img1, img2, 
                    star, 
                    brief,
                    is_binary_descriptor=True, 
                    match_method='BRUTE_FORCE', 
                    description='STAR + BRIEF, BRTUE_FORCE, RANSAC')


fast = cv2.FastFeatureDetector_create()
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
matcher.draw_ransac(img1, img2, 
                    fast, 
                    brief,
                    is_binary_descriptor=True, 
                    match_method='BRUTE_FORCE', 
                    description='FAST + BRIEF, BRTUE_FORCE, RANSAC')
```
