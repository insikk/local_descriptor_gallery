# STAR(CenSurE) Feature and BRIEF (Binary Robust Independent Elementary Features) Descriptor

STAR detector(also known as CenSurE detector)

See https://docs.opencv.org/3.2.0/dc/d7d/tutorial_py_brief.html for more detail

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

img = cv2.imread('pet_bottle.jpg')
# Initiate FAST detector
star = cv2.xfeatures2d.StarDetector_create()
# Initiate BRIEF extractor
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
# find the keypoints with STAR
kp = star.detect(img,None)
# compute the descriptors with BRIEF
kp, des = brief.compute(img, kp)


img2 = cv2.drawKeypoints(img, kp, None, color=(255,0,0))
# Print all default params
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB);
plt.figure(figsize=(16, 12))
plt.imshow(img2)
plt.show()
```
